```python
# sort out 'bathrooms_text' -> split into two columns : # of baths , shared/non-shared

import pandas as pd
import re

# # Sample DataFrame
# data = {'bathrooms_text': ['2 shared baths', '1 private bath', '3.5 baths', 'shared half-bath']}
# df = pd.DataFrame(data)

# Function to extract the number of shared bathrooms
def extract_shared_bathrooms(text):
    if isinstance(text, str):
        if 'shared' in text.lower():
            return 1  # Indicate the presence of shared bathrooms
        else:
            return 0  # Indicate no shared bathrooms
    else:
        return 0  # Indicate no shared bathrooms for non-string values

# Create a new 'bathroom_shared' column
df['bathroom_shared'] = df['bathrooms_text'].apply(extract_shared_bathrooms)

# Function to extract the number of bathrooms as a float
def extract_bathroom_number(text):
    if isinstance(text, str):
        # Check if it's a shared bathroom and remove 'shared'
        text = re.sub(r'shared', '', text, flags=re.IGNORECASE)
        # Extract numbers (including decimals)
        nums = re.findall(r'\d+\.\d+|\d+', text)
        return float(nums[0]) if nums else None
    else:
        return None

# Apply the extraction function to 'bathrooms_text' column
df['bathrooms_text'] = df['bathrooms_text'].apply(extract_bathroom_number)

# Let's see the updated DataFrame 
print(df[['bathrooms_text', 'bathroom_shared']])  #bathroom_shared ==1 -> "shared" 
```

```python
# neighborhood n/a 

zipcodes = df['neighbourhood_cleansed'].unique()
zipcodes

neighborhoods_by_zipcode = df.groupby('neighbourhood_cleansed')['neighbourhood'].unique()

# Convert the Series to a dictionary
zipcode_to_neighborhoods_dict = neighborhoods_by_zipcode.to_dict()

zipcode_to_neighborhoods_dict


final_dict = {}

for zipcode, neighborhoods in zipcode_to_neighborhoods_dict.items():
    # Filter out 'Austin, Texas, United States' and NaN values
    filtered_neighborhoods = [n for n in neighborhoods if n != 'Austin, Texas, United States' and pd.notna(n)]
    
    # If there's at least one valid neighborhood name, use the first one. Otherwise, default to 'Austin, Texas, United States'.
    if filtered_neighborhoods:
        final_dict[zipcode] = filtered_neighborhoods[0]
    else:
        final_dict[zipcode] = 'Austin, Texas, United States'

final_dict
```

```python
# Replace the 'neighbourhood' column values based on the 'neighbourhood_cleansed' mapping
df['neighbourhood'] = df['neighbourhood_cleansed'].map(final_dict)

df['neighbourhood'] = df['neighbourhood'].replace('Austin, Texas, United States', 'Austin Texas, United States')
df['neighbourhood'] = df['neighbourhood'].replace('Austin, United States', 'Austin Texas, United States')

# Now, 'df' will have the 'neighbourhood' column updated with the values from 'final_dict' where the keys match the 'neighbourhood_cleansed' column.
df.info()
```

### consolidating comments for each listing_id into one entry

```python
import pandas as pd

def combine_comments(df):
    # Check if 'comments' column exists in the DataFrame
    if 'comments' in df.columns:
        # Convert 'comments' column to strings
        df['comments'] = df['comments'].astype(str)
    
        # Group by 'id' column and combine 'comments' into a single list
        combined_comments = df.groupby('id')['comments'].apply(list).reset_index()
    
        # Create a new column 'all_comments' to store the combined list of comments
        df['all_comments'] = df['id'].map(combined_comments.set_index('id')['comments'])
    
    return df

```

```python
# Call the function with your DataFrame
combined_df = combine_comments(combined_df)
```

```python
combined_df['all_comments'] = combined_df['all_comments'].apply(lambda x: ', '.join(x))
```

```python
import nltk
import string

# Download NLTK stopwords data if you haven't already
nltk.download('stopwords')

from nltk.corpus import stopwords

def preprocess_text(text):
    # Tokenize the text into words
    words = nltk.word_tokenize(text)
    
    # Remove punctuation, backslashes, and double quotes
    words = [word for word in words if word not in (string.punctuation + '\\\"')]
    
    # Convert to lowercase and remove stopwords
    words = [word.lower() for word in words if word.lower() not in stopwords.words('english')]
    
    return words
```

```python
combined_df['desc_words'] = combined_df['description'].apply(preprocess_text)
combined_df['host_words'] = combined_df['host_about'].apply(preprocess_text)

```

```python
combined_df['all_comments'] = combined_df['all_comments'].apply(preprocess_text)
```
